WHERE clause: select all records where the City column has the value 'Berlin' in customers

```python
%%sql
SELECT *
FROM customers
WHERE City = 'Berlin';
```

AND operator: select all records where the city column has the value 'Dehli' and the postal code column has 110017

```python
%%sql
SELECT *
FROM customers
WHERE City= 'Delhi' AND PostalCode= 110017
```

OR operator: select all records where the city column has the value 'Belin" or 'London'

```python
%%sql
SELECT *
FROM customers
WHERE City = 'Berlin' OR City = 'London';
```

NOT operator: USe the NOT keyword to select all records where country is not Germany

```python
%%sql
SELECT *
FROM customers
WHERE NOT Country = 'Germany';
```

% Wildcard: select all records where the second letter of the city is an 'a'

```python
%%sql
SELECT *
FROM customers
WHERE City LIKE '_a%';
```

ORDER BY: select all records from the customers table, sort the result alphabetically by the column city

```python
%%sql
SELECT *
FROM customers
ORDER BY City ASC;
```

Find all the tracks that have length 5,000,000 milliseconds or more

```python
%%sql
SELECT *
FROM tracks
WHERE milliseconds >= 5000000
```

Find all the invoices whose total is between $10 and $20

```python
%%sql
SELECT *
FROM invoices
WHERE total BETWEEN 10 AND 20
```

Find all the customers from the following states: RJ, DF, AB, BC, CA, WA, NY

```python
%%sql
SELECT *
FROM customers
WHERE State IN ('RJ', 'DF', 'AB', 'BC', 'CA', 'WA', 'NY');
```

Find all the invoices for customer 56 and 58 where the total was between $1 and $5

```python
%%sql
SELECT *
FROM invoices
WHERE total BETWEEN 1 AND 5 AND (customerID = '58' OR customerID = '56')
```

 Find all the customer emails that start with “J” and are from gmail.com

```python
%%sql
SELECT *
FROM customers
WHERE email LIKE 'J%@gmail.com'
```
